```python
from dotenv import load_dotenv
load_dotenv()
import os
import pandas as pd

client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")

import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

spotify =spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id = client_id,
                                                              client_secret = client_secret))
##Imagine Dragons
id= "53XhwfbYqKCa1cC15pYq2q"

top_tracks = spotify.artist_top_tracks(id)['tracks']

song_data = [{'Nombre': track['name'], 'Popularidad': track['popularity'], 'Duracion (min)': track['duration_ms'] / (60 * 1000)} for track in top_tracks]

df = pd.DataFrame(song_data)
df = df.sort_values(by='Popularidad', ascending=True)
print(df.head(3))
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['Duracion (min)'], df['Popularidad'])
plt.title('Relación entre Duración y Popularidad de Canciones')
plt.xlabel('Duración (minutos)')
plt.ylabel('Popularidad')
plt.show()
```

Does duration have a relationship with popularity? Could we say that a song that lasts a short time may be more popular than a song that lasts longer?
